```python
import json
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np


location = pd.read_excel(r'C:\Users\joaki\Google Drive\dnd\The Broom\broom of Faerun v2.xlsx',sheet_name='location', delimiter = ';')
location['Location2'] = location['Location2'].fillna('')
location['Location_description'] = location['Location_description'].fillna('')
location['Location_name']=(""+location['Location1']+" "+location['Location2']+"").str.replace(" ","_")
location['Main_location_name']=(""+location['Location1']+"").str.replace(" ","_")
location['Location_nr']=location.groupby('Main_location_name').ngroup(ascending=True)+1
location=location.sort_values(by = 'Main_location_name')
location['Location1_nr'] = location.groupby('Main_location_name').cumcount()+1
location['Row'] = np.arange(len(location))+1
location.to_json(r'C:\Users\joaki\Google Drive\dnd\The Broom\locations.json',orient='records')
location.to_json(r'C:\Users\joaki\Documents\GitHub\Broom\locations.json',orient='records')


quest = pd.read_excel(r'C:\Users\joaki\Google Drive\dnd\The Broom\broom of Faerun v2.xlsx',sheet_name='quest', delimiter = ';')
quest['Quest_giver'] = quest['Quest_giver'].fillna('')
quest['Quest_target'] = quest['Quest_target'].fillna('')
quest['Location1'] = quest['Location1'].fillna('')
quest['Location2'] = quest['Location2'].fillna('')
quest['Description'] = quest['Description'].fillna('')
quest['Reward'] = quest['Reward'].fillna('')
quest['Main_location_name']=(""+quest['Location1']+"").str.replace(" ","_")
quest['Location_name']=(""+quest['Location1']+" "+quest['Location2']+"").str.replace(" ","_")
quest.to_json(r'C:\Users\joaki\Google Drive\dnd\The Broom\quest.json',orient='records')
quest.to_json(r'C:\Users\joaki\Documents\GitHub\Broom\quest.json',orient='records')

person = pd.read_excel(r'C:\Users\joaki\Google Drive\dnd\The Broom\broom of Faerun v2.xlsx',sheet_name='person', delimiter = ';')
person['Location2'] = person['Location2'].fillna('')
person['Main_location_name']=(""+person['Location1']+" "+person['Location2']+"").str.replace(" ","_")
person=person.sort_values(by = ['Main_location_name','First_Name'])
person['Location_nr']=person.groupby('Main_location_name').ngroup(ascending=True)+1
person['Location1_nr'] = person.groupby('Location1').cumcount()+1
person['Location2_nr'] = person.groupby('Main_location_name').cumcount()+1
person['Row'] = (np.arange(len(person)))
person['Name'] = person['Name'].fillna('')
person['First_Name'] = person['First_Name'].fillna('')
person['Id2']='ID'
person['ID']=(""+person['Id2']+""+person['First_Name']+"")
person=person.sort_values(by = 'Column')
person['test']=person.groupby('Column').cumcount()+1
person['First_faction']= person['test'].apply(lambda x: 1 if x==1 
                                                        else 0)
person=person.sort_values(by = ['Main_location_name','First_Name'])
person.drop(['test'], axis='columns', inplace=True)

quest2=quest[quest["Completed"]==0]
Quest_giver=pd.DataFrame(quest2['Quest_giver'].unique())
Quest_giver['Quest_giver']=1
Quest_giver = Quest_giver.rename(columns = {0: 'Quest_giver_name'}, inplace = False)
Quest_target=pd.DataFrame(quest2['Quest_target'].unique())
Quest_target['Quest_target']=1
Quest_target = Quest_target.rename(columns = {0: 'Quest_target_name'}, inplace = False)

person = pd.merge(person, Quest_target, how='left', left_on='First_Name', right_on='Quest_target_name')
person.drop(['Quest_target_name'], axis='columns', inplace=True)
person = pd.merge(person, Quest_giver, how='left', left_on='First_Name', right_on='Quest_giver_name')
person.drop(['Quest_giver_name'], axis='columns', inplace=True)
person['Quest_target'] = person['Quest_target'].fillna('0')
person['Quest_giver'] = person['Quest_giver'].fillna('0')
person=person.sort_values(by = ['Main_location_name','First_Name'])
person.to_json(r'C:\Users\joaki\Google Drive\dnd\The Broom\person.json',orient='records')
person.to_json(r'C:\Users\joaki\Documents\GitHub\Broom\person.json',orient='records')

treasury = pd.read_excel(r'C:\Users\joaki\Google Drive\dnd\The Broom\broom of Faerun v2.xlsx',sheet_name='treasury', delimiter = ';')
treasury= treasury.fillna('')
treasury.to_json(r'C:\Users\joaki\Google Drive\dnd\The Broom\treasury.json',orient='records')
treasury.to_json(r'C:\Users\joaki\Documents\GitHub\Broom\treasury.json',orient='records')

magic_items= pd.read_excel(r'C:\Users\joaki\Google Drive\dnd\The Broom\broom of Faerun v2.xlsx',sheet_name='Shared magical items', delimiter = ';')
magic_items.to_json(r'C:\Users\joaki\Google Drive\dnd\The Broom\magic_items.json',orient='records')
magic_items.to_json(r'C:\Users\joaki\Documents\GitHub\Broom\magic_items.json',orient='records')

legend= pd.read_excel(r'C:\Users\joaki\Google Drive\dnd\The Broom\broom of Faerun v2.xlsx',sheet_name='legend', delimiter = ';')
legend.to_json(r'C:\Users\joaki\Google Drive\dnd\The Broom\legend.json',orient='records')
legend.to_json(r'C:\Users\joaki\Documents\GitHub\Broom\legend.json',orient='records')
```
